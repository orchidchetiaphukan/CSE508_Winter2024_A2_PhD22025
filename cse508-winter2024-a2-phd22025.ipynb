# Image Feature Extraction

```python
import cv2
import os

def process_image(image_path, output_dir):
    img = cv2.imread(image_path)
    
    # Resizing the image
    resized_img = cv2.resize(img, (new_width, new_height))
    
    # Adjusting brightness
    brightness = 50
    adjusted_img = cv2.convertScaleAbs(resized_img, beta=brightness)
    
    # Adjusting exposure
    exposure = 1.5
    adjusted_img = cv2.convertScaleAbs(adjusted_img, alpha=exposure)
    
    filename = os.path.basename(image_path)
    
 
    output_path = os.path.join(output_dir, filename)
    

    cv2.imwrite(output_path, adjusted_img)


new_width = 224
new_height = 224

output_directory = '/kaggle/working/processed_images'


if not os.path.exists(output_directory):
    os.makedirs(output_directory)

for index, row in images_df.iterrows():
    image_path = row['file_path']
    process_image(image_path, output_directory)
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image

from keras.models import Model
```

```python

base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

global_avg_pool_output = base_model.layers[-1].output


resnet_model = Model(inputs=base_model.input, outputs=global_avg_pool_output)


def extract_resnet_features(image_path):

    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    features = resnet_model.predict(img_array)
    features = np.mean(features, axis=(1, 2))
    normalized_features = features / np.linalg.norm(features)
    
    return normalized_features


image_path = df_final_final['file_path_y'][3] 
mean_features = extract_resnet_features(image_path)
print("Mean of extracted features:", mean_features.shape)
```

# Text Feature Extraction

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
nltk.download('wordnet')
nltk.download('stopwords')
def cleaning(data):
    stemmer = PorterStemmer()
    corpus = []
    for i in range(0, len(data)):
        sentence = re.sub('[^a-zA-Z]', ' ', str(data[i])) # Removing Punctuation Marks
        sentence = sentence.lower() # Lowering the text
        sentence = sentence.split() # Tokenization
        
        # Remove blank space tokens
        sentence = [word for word in sentence if word.strip()]  # Removes empty strings
        
        all_stopwords = stopwords.words('english') # Removing the stopwords
        #all_stopwords.remove('not')
        
        sentence = [stemmer.stem(word) for word in sentence if not word in set(all_stopwords)]
        sentence = ' '.join(sentence)
        corpus.append(sentence)
      
    return corpus
```

```python
df_final_final['preprocessed_text'] = cleaning(df_final_final['Review Text'])
df_final_final.head()
```

```python
from collections import Counter
import math

def calculate_tf_idf_for_corpus(corpus):
    # Tokenization
    tokenized_corpus = [document.split() for document in corpus]
    
    # Calculating TF
    tf_list = []
    for doc_tokens in tokenized_corpus:
        tf = Counter(doc_tokens)
        tf_list.append(tf)
    
    # Calculating DF
    df = Counter()
    for tf in tf_list:
        df.update(tf.keys())
    
    # Calculating IDF
    num_docs = len(corpus)
    idf = {word: math.log(num_docs / df[word]) for word in df}
    
    # Calculating TF-IDF scores
    tf_idf_list = []
    for tf in tf_list:
        tf_idf = {word: tf[word] * idf[word] for word in tf}
        tf_idf_list.append(tf_idf)
    
    return tf_idf_list


corpus = df_final_final['preprocessed_text'].tolist()
tf_idf_scores = calculate_tf_idf_for_corpus(corpus)

df_tf_idf = pd.DataFrame(tf_idf_scores)
df_tf_idf = df_tf_idf.fillna(0)
print(df_tf_idf.shape)
df_tf_idf
```

# 3. Image Retrieval and Text Retrieval

## a.

```python

def cosine_similarity(vec1, vec2):
    """Calculate cosine similarity between two vectors."""
    dot_prod = np.dot(vec1, vec2)
    mag_vec1 = np.linalg.norm(vec1)
    mag_vec2 = np.linalg.norm(vec2)
    
    if mag_vec1 != 0 and mag_vec2 != 0:
        return dot_prod / (mag_vec1 * mag_vec2)
    else:
        return 0  # To handle division by zero if any vector has zero magnitude

def find_similar_images(input_image_feature, input_review_feature,image_features, threshold=0.8):
    """Find indices of similar images to the input image based on cosine similarity."""
    similarities = np.array([cosine_similarity(input_image_feature, img_feat) for img_feat in image_features])
    similar_indices = np.where(similarities >= threshold)[0]
    similar_cosine_similarities = np.round(similarities[similar_indices], 2)
    return similar_indices, similar_cosine_similarities

similar_indices = find_similar_images(image_features_array[2], text_features_array[2], image_features_array, threshold=0.8)

print("Indices of similar images:", similar_indices)
```

## b.

```python
def find_similar_reviews(input_image_feature, input_review_feature, text_features, threshold=0.8):
    """Find indices of similar reivews to the input image, review pair based on cosine similarity."""
    similarities = np.array([cosine_similarity(input_review_feature, text_feat) for text_feat in text_features])
    similar_indices = np.where(similarities >= threshold)[0]
    similar_cosine_similarities = np.round(similarities[similar_indices], 2)
    return similar_indices, similar_cosine_similarities


similar_indices = find_similar_reviews(image_features_array[2], text_features_array[2], text_features_array, threshold=0.8)

print("Indices of similar reviews:", similar_indices)
```

```python
def find_similar_pairs(input_image_feature, input_review_feature, image_features, text_features, threshold=0.8):
    """Find indices of similar pairs based on composite similarity."""
    image_similarities = np.array([cosine_similarity(input_image_feature, img_feat) for img_feat in image_features])
    text_similarities = np.array([cosine_similarity(input_review_feature, text_feat) for text_feat in text_features])
    
    composite_similarities = (image_similarities + text_similarities) / 2.0
    
    similar_indices = np.where(composite_similarities >= threshold)[0]
    similar_cosine_similarities = np.round(composite_similarities[similar_indices], 2)
    return similar_indices, similar_cosine_similarities

similar_indices = find_similar_pairs(image_features_array[2], text_features_array[2], image_features_array, text_features_array, threshold=0.8)

print("Indices of similar pairs:", similar_indices)
```
